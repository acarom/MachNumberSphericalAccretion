```python
import matplotlib.pyplot as plt
#plt.style.use('seaborn-white')
import numpy as np
```

```python
def h(x, y, gm=3/2):
    f = ((1/(gm-1))+y/2)/(y**((gm-1)/(gm+1)))
    g = ((1/(gm-1))-(3/2)+(2/x))*x**(4*(gm-1)/(gm+1))
    return g/f
```

```python
gamma = 4/3.0
x = np.linspace(0.1, 3, 100)
y = np.linspace(0.1, 4, 100)

X, Y = np.meshgrid(x, y)
Z = h(X, Y, gm=gamma)**((gamma-1)/(gamma+1))
ZZ = Z.copy()
ZZ[ZZ>1.04] = None
```

```python
#%matplotlib widget
plt.subplots(figsize=(12,10), dpi=72)
plt.contourf(X, Y, Z, 16, cmap='PuBu')
#plt.colorbar()
contours = plt.contour(X, Y, ZZ, 12, colors='black')
plt.clabel(contours, inline=True, fontsize=14*1.5, manual= [(1,2),(2,1),(1,0.5),(0.5,1), (1,2.5), (1, 3.5), (1.0, 0.29),(0.35, 1),(1.5, 1),(2.6, 1),(2.9,1)])

contours = plt.contour(X, Y, ZZ, 1, colors='black', linewidths= 3)
plt.clabel(contours, inline=True, fontsize=14*1.5, manual= [(1.6,1.6),(1.25,0.7)])
plt.yticks(fontsize=18)
plt.xticks(fontsize=18)
plt.xlabel(r'$r/r_s$', fontsize=18*1.5)
plt.ylabel(r'$M^2 = (v / c_s)^2$', fontsize=18*1.5)

plt.text(0.96, 2.19, r'$\mathbf{4}$',c='white', fontsize=18*1.5)
plt.text(0.97, 2.2, r'$\mathbf{4}$', fontsize=18*1.5)

plt.text(0.96, 0.64, r'$\mathbf{3}$',c='white', fontsize=18*1.5)
plt.text(0.97, 0.65, r'$\mathbf{3}$', fontsize=18*1.5)

plt.text(0.48, 1.44, r'$\mathbf{6}$',c='white', fontsize=18*1.5)
plt.text(0.49, 1.45, r'$\mathbf{6}$', fontsize=18*1.5)

plt.text(2.28, 1.44, r'$\mathbf{5}$',c='white', fontsize=18*1.5)
plt.text(2.29, 1.45, r'$\mathbf{5}$', fontsize=18*1.5)

plt.text( 0.76, 1.25, r'$\mathbf{1}$',c='white', fontsize=18*1.5)
plt.text( 0.77, 1.265, r'$\mathbf{1}$', fontsize=18*1.5)

plt.text( 1.16, 1.25, r'$\mathbf{2}$',c='white', fontsize=18*1.5)
plt.text( 1.17, 1.265, r'$\mathbf{2}$', fontsize=18*1.5)

plt.savefig('AccretionSphericallySym.pdf')  
```
